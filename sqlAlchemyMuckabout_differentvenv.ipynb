```python
cards_5 = []
for text in ResultSet:
    cards_5.append(text[0].replace('\n',' ').strip())
    ['on', 'the', 'of', 'to', 'and', 'with', 'in']
cards_5 = ' '.join(cards_5)
```

```python
pat = re.compile(['on','the'])
```

```python
# re.findall(r'on|the|of|to|and|with|in', cards_5)
```

```python
cards_5 = re.sub(r' on | the | of | to | and | with | in ',' ',cards_5)
```

```python
npl = spacy.load("en_core_web_sm")
```

```python
doc = npl(cards_5)
```

```python
lemmas = []
for token in doc:
    if not token.is_punct:
        lemmas.append(token.lemma_)
```

```python
lemmas
```

```python
lemetized = ' '.join(lemmas)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
v = CountVectorizer(ngram_range=(2,3))
```

```python
v.fit([lemetized])
```

```python
vocab = v.vocabulary_
```

```python
for phrase in vocab:
    count = cards_5.count(phrase)
    if count > 25 and count < 200:
        print(phrase+': ', count)
```
